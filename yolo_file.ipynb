```python
#YOLO sample 


from ultralytics import YOLO
model = YOLO("yolov8n.pt")  
results = model("chairs.jpg", show=True)
for r in results:
    for box in r.boxes:
        print(
            "Class:", r.names[int(box.cls)],
            "Confidence:", float(box.conf),
            "x, y cordinates", box.xyxy   # x1,y1,x2,y2
        )
```

```python
from ultralytics import YOLO
from collections import Counter

# Load YOLO model
model = YOLO("yolov8n.pt")

# Run inference
results = model("peoples.jpg", show=True)

object_list = []

# Collect detected object names
for r in results:
    for box in r.boxes:
        class_name = r.names[int(box.cls)]
        object_list.append(class_name)
        print(
            "Class:", r.names[int(box.cls)],
            "Confidence:", float(box.conf),
            "x, y cordinates", box.xyxy   # x1,y1,x2,y2
        )

# Count objects
object_counts = Counter(object_list)

# Print results
print("\nDetected Objects and Counts:")
for obj, count in object_counts.items():
    print(f"{obj}: {count}")
```
