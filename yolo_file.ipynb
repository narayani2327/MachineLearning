```python
#YOLO sample 


from ultralytics import YOLO
model = YOLO("yolov8n.pt")  
results = model("cat.jpeg", show=True)
for r in results:
    for box in r.boxes:
        print(
            "Class:", r.names[int(box.cls)],
            "Confidence:", float(box.conf),
            "x, y cordinates", box.xyxy   # x1,y1,x2,y2
        )
```
