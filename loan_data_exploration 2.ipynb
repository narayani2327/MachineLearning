```python
# Fill null values in 'Married' column with mode (most frequent value)

# Check if column exists
if 'Married' in df.columns:
    # Count nulls before
    null_before = df['Married'].isnull().sum()
    print(f"Null values in 'Married' before filling: {null_before}")
    
    # Find mode (most frequent value)
    married_mode = df['Married'].mode()[0]
    print(f"Mode of 'Married' column: {married_mode}")
    
    # Fill null values with mode
    df['Married'].fillna(married_mode, inplace=True)
    
    # Count nulls after
    null_after = df['Married'].isnull().sum()
    print(f"Null values in 'Married' after filling: {null_after}")
    
    # Show value counts
    print("\nValue counts in 'Married' column:")
    print(df['Married'].value_counts())
else:
    print("Column 'Married' not found in dataset!")
    print(f"Available columns: {df.columns.tolist()}")
```

```python
# Analyze null columns and suggest best filling method
print("="*80)
print("RECOMMENDED FILLING STRATEGIES FOR NULL COLUMNS")
print("="*80)

null_cols = df.columns[df.isnull().any()].tolist()

if len(null_cols) > 0:
    for col in null_cols:
        null_count = df[col].isnull().sum()
        null_pct = (null_count / len(df)) * 100
        dtype = df[col].dtype
        
        print(f"\n Column: {col}")
        print(f"   Type: {dtype}")
        print(f"   Nulls: {null_count} ({null_pct:.2f}%)")
        
        # Suggest filling method based on type and characteristics
        if pd.api.types.is_numeric_dtype(df[col]):
            # Check for outliers using IQR
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            outliers = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))][col].count()
            
            if outliers > len(df) * 0.05:  # More than 5% outliers
                print(f"    RECOMMENDED: MEDIAN (has {outliers} outliers)")
            else:
                print(f"    RECOMMENDED: MEAN (no significant outliers)")
            print(f"    Mean: {df[col].mean():.2f}, Median: {df[col].median():.2f}")
        else:
            # Categorical column
            mode_val = df[col].mode()[0] if len(df[col].mode()) > 0 else "N/A"
            unique_count = df[col].nunique()
            print(f"    RECOMMENDED: MODE (most frequent value)")
            print(f"    Mode: {mode_val}, Unique values: {unique_count}")
            print(f"   Alternative: 'Unknown' or 'Not Available'")
else:
    print("No null columns found!")
```

```python
# Apply recommended filling strategies
print("Filling null values with recommended strategies...\n")

# Categorical columns - fill with MODE
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
print(f" Gender filled with mode: {df['Gender'].mode()[0]}")

df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
print(f" Dependents filled with mode: {df['Dependents'].mode()[0]}")

df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
print(f" Self_Employed filled with mode: {df['Self_Employed'].mode()[0]}")

# Numerical columns - fill with MEDIAN (due to outliers)
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
print(f" LoanAmount filled with median: {df['LoanAmount'].median()}")

df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median())
print(f" Loan_Amount_Term filled with median: {df['Loan_Amount_Term'].median()}")

df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].median())
print(f" Credit_History filled with median: {df['Credit_History'].median()}")

# Loan_Status - fill with mode
df['Loan_Status'] = df['Loan_Status'].fillna(df['Loan_Status'].mode()[0])
print(f" Loan_Status filled with mode: {df['Loan_Status'].mode()[0]}")

# Verify no nulls remain
print("\n" + "="*60)
print("VERIFICATION: Remaining null values")
print("="*60)
remaining_nulls = df.isnull().sum()
if remaining_nulls.sum() == 0:
    print(" SUCCESS: All null values have been filled!")
else:
    print(remaining_nulls[remaining_nulls > 0])
```

```python
# Check if target column (Loan_Status) is balanced or imbalanced
print("="*70)
print("TARGET COLUMN BALANCE CHECK: Loan_Status")
print("="*70)

# Count and percentage distribution
value_counts = df['Loan_Status'].value_counts()
value_percentage = df['Loan_Status'].value_counts(normalize=True) * 100

print("\nValue Counts:")
print(value_counts)
print("\nPercentage Distribution:")
print(value_percentage)

# Calculate balance ratio
total = value_counts.sum()
majority_class = value_counts.max()
minority_class = value_counts.min()
balance_ratio = minority_class / majority_class

print(f"\nTotal samples: {total}")
print(f"Majority class: {majority_class} ({(majority_class/total)*100:.2f}%)")
print(f"Minority class: {minority_class} ({(minority_class/total)*100:.2f}%)")
print(f"Balance ratio: {balance_ratio:.2f}")

# Determine if balanced or imbalanced
print("\n" + "="*70)
if balance_ratio >= 0.8:
    print(" BALANCED: Dataset is well balanced (ratio >= 0.8)")
elif balance_ratio >= 0.6:
    print(" SLIGHTLY IMBALANCED: Dataset is slightly imbalanced (0.6 <= ratio < 0.8)")
else:
    print(" IMBALANCED: Dataset is imbalanced (ratio < 0.6)")
    print("   Consider using techniques like SMOTE, undersampling, or class weights")
print("="*70)

# Visualize the distribution
plt.figure(figsize=(10, 5))

# Bar plot
plt.subplot(1, 2, 1)
value_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Loan_Status Distribution (Count)', fontsize=12, fontweight='bold')
plt.xlabel('Loan Status', fontsize=10)
plt.ylabel('Count', fontsize=10)
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)

# Pie chart
plt.subplot(1, 2, 2)
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', 
        colors=['skyblue', 'lightcoral'], startangle=90)
plt.title('Loan_Status Distribution (Percentage)', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Apply SMOTE to balance the dataset
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder

print("="*70)
print("APPLYING SMOTE TO BALANCE THE DATASET")
print("="*70)

# Prepare data: Encode categorical variables for SMOTE
# SMOTE requires numerical data, so we need to encode categorical columns

# Create a copy of the dataframe for encoding
df_encoded = df.copy()

# Store label encoders for later reference
label_encoders = {}

# Encode all categorical columns
categorical_columns = df_encoded.select_dtypes(include=['object']).columns.tolist()

print("\nEncoding categorical columns...")
for col in categorical_columns:
    if col != 'Loan_Status':  # Don't encode target yet
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df_encoded[col])
        label_encoders[col] = le
        print(f"   {col} encoded")

# Separate features and target
X = df_encoded.drop('Loan_Status', axis=1)
y = df_encoded['Loan_Status']

# Encode target variable
le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y)

print(f"\nOriginal dataset shape: {X.shape}")
print(f"Original class distribution:")
print(f"  Y (Approved): {(y == 'Y').sum()}")
print(f"  N (Not Approved): {(y == 'N').sum()}")

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y_encoded)

print(f"\nAfter SMOTE:")
print(f"Resampled dataset shape: {X_resampled.shape}")
print(f"Resampled class distribution:")
unique, counts = np.unique(y_resampled, return_counts=True)
for val, count in zip(unique, counts):
    label = le_target.inverse_transform([val])[0]
    print(f"  {label}: {count}")

# Create balanced dataframe
df_balanced = pd.DataFrame(X_resampled, columns=X.columns)
df_balanced['Loan_Status'] = le_target.inverse_transform(y_resampled)

# Decode categorical columns back to original values
print("\nDecoding categorical columns back...")
for col in label_encoders.keys():
    # Round to nearest integer and convert to int for decoding
    df_balanced[col] = df_balanced[col].round().astype(int)
    # Ensure values are within valid range
    df_balanced[col] = df_balanced[col].clip(0, len(label_encoders[col].classes_) - 1)
    df_balanced[col] = label_encoders[col].inverse_transform(df_balanced[col])
    print(f"   {col} decoded")

print("\n" + "="*70)
print(" SMOTE APPLIED SUCCESSFULLY!")
print(f"Final balanced dataset shape: {df_balanced.shape}")
print("="*70)
```

```python
# Visualize the before and after SMOTE comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Before SMOTE
original_counts = df['Loan_Status'].value_counts()
axes[0].bar(original_counts.index, original_counts.values, color=['skyblue', 'lightcoral'])
axes[0].set_title('Before SMOTE - Imbalanced Dataset', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Loan Status', fontsize=11)
axes[0].set_ylabel('Count', fontsize=11)
axes[0].set_ylim(0, 850)
axes[0].grid(axis='y', alpha=0.3)
for i, v in enumerate(original_counts.values):
    axes[0].text(i, v + 20, str(v), ha='center', fontweight='bold')

# After SMOTE
balanced_counts = df_balanced['Loan_Status'].value_counts()
axes[1].bar(balanced_counts.index, balanced_counts.values, color=['skyblue', 'lightcoral'])
axes[1].set_title('After SMOTE - Balanced Dataset', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Loan Status', fontsize=11)
axes[1].set_ylabel('Count', fontsize=11)
axes[1].set_ylim(0, 850)
axes[1].grid(axis='y', alpha=0.3)
for i, v in enumerate(balanced_counts.values):
    axes[1].text(i, v + 20, str(v), ha='center', fontweight='bold')

plt.tight_layout()
plt.show()

print("\n" + "="*70)
print("SUMMARY:")
print("="*70)
print(f"Original dataset: {df.shape[0]} samples (Y: 789, N: 192)")
print(f"Balanced dataset: {df_balanced.shape[0]} samples (Y: 789, N: 789)")
print(f"Synthetic samples created: {df_balanced.shape[0] - df.shape[0]}")
print("="*70)
```
